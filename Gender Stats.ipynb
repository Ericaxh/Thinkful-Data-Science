```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('genderstats.csv')
print(df)
```

```python
Countries = df['Country Name'][:30]
y2010 = df['2010 [YR2010]']
y2011 = df['2011 [YR2011]']
y2012 = df['2012 [YR2012]']
y2013 = df['2013 [YR2013]']
y2014 = df['2014 [YR2014]']
y2015 = df['2015 [YR2015]']
y2016 = df['2016 [YR2016]']
y2017 = df['2017 [YR2017]']
y2018 = df['2018 [YR2018]']
years = [y2010, y2011, y2012, y2013, y2014, y2015, y2016, y2017, y2018]
```

```python
year_array = np.array([y2010[:30], y2011[:30], y2012[:30], y2013[:30], y2014[:30], y2015[:30], y2016[:30], y2017[:30], y2018[:30]])
years = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018])
gender_emp = pd.DataFrame(year_array)
gender_emp.index = years
gender_emp.columns = Countries
gender_emp
```

```python
plt.plot(gender_emp['Switzerland'], color='purple')
plt.show()
```

```python
plt.plot(gender_emp['India'], color='green')
plt.show()
```

```python
plt.plot(gender_emp['United Kingdom'], color='purple')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(gender_emp['Switzerland'], color='purple')
plt.xlabel("Year")
plt.ylabel("WIW")
plt.title("Switzerland")

plt.subplot(2, 2, 2)
plt.plot(gender_emp['Denmark'], color='blue')
plt.xlabel("Year")
plt.ylabel("WIW")
plt.title("Denmark")

plt.subplot(2, 2, 3)
plt.plot(gender_emp['Canada'], color='green')
plt.xlabel("Year")
plt.ylabel("WIW")
plt.title("Canada")

plt.subplot(2, 2, 4)
plt.plot(gender_emp['France'], color='orange')
plt.xlabel("Year")
plt.ylabel("WIW")
plt.title("France")

plt.tight_layout()
```

```python
from the data it appears that all of the grapped countries listed have had a strong and steady increase in women in the workforce. However,some locations like India experienced a significant dip for certain years which may be explained by other factors. The next step would be to analyze the numbers of men and women in the workforce relative to the total population to make any further analysis.
```
